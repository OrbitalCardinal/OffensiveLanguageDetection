In this section exploratory data analysis is made to the clean datasets.

We can learn and gain useful insights on how the data is, and check if the data makes sense in order to ensure it is a good training set for a machine learning algorithm.

Reading picles of document term matrix

```python
tweets_dtm = pd.read_pickle('dtm.pkl')
not_tweets_dtm = pd.read_pickle('dtm_not.pkl')
off_tweets_dtm = pd.read_pickle('dtm_off.pkl')
```

Reading picles of corpus

```python
corpus_off = pd.read_pickle('corpus_off.pkl')
corpus_not = pd.read_pickle('corpus_not.pkl')
corpus_off
```

Making offensive and not offensive tweets into a single string for wordclouds

```python
offensive_wc = ""
not_offensive_wc = ""
for index, tweet in enumerate(corpus_off.transpose().columns):
    offensive_wc += corpus_off.tweet[tweet]
for index, tweet in enumerate(corpus_not.transpose().columns):
    not_offensive_wc += corpus_not.tweet[tweet]
```

### Plotting the top ten most frequent words

```python
from wordcloud import WordCloud
from sklearn.feature_extraction import text

stop_words = text.ENGLISH_STOP_WORDS
wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
              max_font_size=150,random_state=42)

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,2,1)
wc.generate(offensive_wc)
plt.axis("off")
plt.title("Offensive")
plt.imshow(wc,interpolation="bilinear")
ax = fig.add_subplot(1,2,2)
wc.generate(not_offensive_wc)
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
plt.title("Not Offensive")
# axes[1].set_title("Not offensive")
plt.show()
```

## Document term matrix

Scikit-learn is used to get the document term matrix from the previous cleaned data.
We use this to get numerical data on the frequency of the words in offensive and not offensive tweets.

```python
only_tweets = pd.read_csv("only_tweets.csv")
only_tweets_off = pd.read_csv("only_tweets_off.csv")
only_tweets_not = pd.read_csv("only_tweets_not.csv")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
tweets_cv = cv.fit_transform(only_tweets.tweet)
tweets_dtm = pd.DataFrame(tweets_cv.toarray(), columns=cv.get_feature_names())
tweets_dtm

off_cv = CountVectorizer(stop_words='english')
off_tweets_cv = off_cv.fit_transform(only_tweets_off.tweet)
off_tweets_dtm = pd.DataFrame(off_tweets_cv.toarray(), columns=off_cv.get_feature_names())

not_cv = CountVectorizer(stop_words='english')
not_tweets_cv = not_cv.fit_transform(only_tweets_not.tweet)
not_tweets_dtm = pd.DataFrame(not_tweets_cv.toarray(), columns=not_cv.get_feature_names())

tweets_dtm
```

A function is made in order to perform a vertical sum on columns and get the frequency of a word in every offensive and not offensive tweet.

```python
def word_counting(dtm):
    word_count = dtm.sum()
    word_count = word_count.sort_values()[::-1]
    word_count = pd.DataFrame(word_count)
    word_count.columns = ["frequency"]
    return word_count
```

Getting the frequency of the words

```python
word_count_not = word_counting(not_tweets_dtm)
word_count_off = word_counting(off_tweets_dtm)
word_count = word_counting(tweets_dtm)
```
